```python
# Import dependencies.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the WeatherPy_vacation.csv file from Part 1 as a new DataFrame.
cities_vacation_df = pd.read_csv("../Weather_Database/WeatherPy_database.csv")
cities_vacation_df.head()
```

```python
#Prompt the customer for the minimum  and maximum temperature preference, rain and snow.
print("What is your minimum temperature preference?")
min_temp = float(input("min_temp: "))

print("What is your maximum temperature preference?")
max_temp = float(input("max_temp: "))
```

```python
preffered_cities_df = cities_vacation_df.loc[(cities_vacation_df["Max Temp"] > min_temp) & (cities_vacation_df["Max Temp"] < max_temp)]

preffered_cities_df.head()
```

```python
weather_vacation_df = preffered_cities_df[["City","Country","Max Temp","Current Description","Lat","Lng"]].copy()
weather_vacation_df.head()
```

```python
weather_vacation_df["Hotel Name"] = ""
weather_vacation_df.head()
```

```python
#Import requests
import requests


from config import g_key
```

```python
# Search for hotels and add them in the weather dataframe.
for i, row in preffered_cities_df[51:54].iterrows():
    
    lat = row["Lat"]
    lng = row["Lng"]
    location = f"{lat},{lng}"
   
    params = {"radius" : 5000,
         "key": g_key,
         "type": "lodging",
        "location": f"{lat},{lng}"}
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    hotels = requests.get(base_url, params=params).json()
#     print(hotels)
    try:
        weather_vacation_df.loc[i,"Hotel Name"] = hotels["results"][0]["name"]
    except:
        print("Hotel not found....skipping.")
    
```

```python
#weather_vacation_df.to_csv("data/WeatherPy_vacation.csv", index_label="City ID")

weather_vacation_df = pd.read_csv("WeatherPy_vacation-backup.csv")
weather_vacation_df.head(10)
```

```python
import gmaps
#Adding markers.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in weather_vacation_df.iterrows()]

#Create the heatmap of temperature for the vacation spots and marker for each city.
locations = weather_vacation_df[["Lat","Lng"]]
# max_temp = weather_vacation_df["Max Temp"]
fig=gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
# heat_layer=gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False, max_intensity=300, point_radius=4)
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
# fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

#Call the figure to plot the data.
fig
```
